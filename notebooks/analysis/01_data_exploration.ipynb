```python
def contem_numeros(celula):
    if isinstance(celula, str):
        return bool(re.search(r'\d', celula))
    return False

celulas_com_numeros = dict_vigitel['Variable name'].apply(contem_numeros)

lista_celulas_com_numeros = [(dict_vigitel.at[index, 'Variable label'], label) 
               for index, label in dict_vigitel[celulas_com_numeros]['Variable name'].items()]

lista_celulas_com_numeros
```
